```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head(4))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop unnamed column
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
df.head()
```

```python
# change weelkly study hourse column
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head(10)
```

```python
# gender distribution
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

## from the above chart we can analyse that:
## the number of girls in the data is more than the number of boys

```python
#grouping the Maths, Reading, Writing scores on the basis of Parents's Education
gb1 = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot= True)
plt.title("Parents Educations and Students average marks")
plt.show()
```

# from the above chart we can analyze that:
# Parent's Education has an considerate impact on the marks of the student

```python
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore": "mean"})
print(gb2)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb2, annot= True)
plt.title("Parents marotal status and Students average marks")
plt.show()
```

# from the above chart we can analyze that:
# Parent Marital Status has negligible impact on the scores of the student

```python
# Distribution of ethnic group in a pie chart by percentage

groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.title("Distribution of Ethnic Group by Percentage")

plt.pie(mlist, labels = l, autopct = "%1.1f%%")
plt.show()

```
